```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# CSV files import
arcs = pd.read_csv('NewArcs.csv')
nodes = pd.read_csv('Nodes.csv')
segments = pd.read_csv('NewLines.csv')

def plot_arc(arc):
    theta = np.linspace(arc['angleStart'], arc['angleEnd'], 100)
    x = arc['x'] + arc['radius'] * np.cos(theta)
    y = arc['y'] + arc['radius'] * np.sin(theta)
    if arc['Rotation'] == 'CCW ':
        plt.plot(x, y, 'g-')  # Use green color for counter-clockwise
    else:
        plt.plot(x, y, 'b-')  # Use blue color for clockwise

def plot_segment(segment):
    if segment['direction'] == '+X':
        dx = -segment['length']
        dy = 0
    elif segment['direction'] == '-X':
        dx = segment['length']
        dy = 0
    elif segment['direction'] == '+Y':
        dx = 0
        dy = -segment['length']
    elif segment['direction'] == '-Y':
        dx = 0
        dy = segment['length']
    x = [segment['x'] + dx, segment['x']]
    y = [segment['y'] + dy, segment['y']]
    plt.plot(x, y, 'k-')  # Use black color for segments

def plot_node(node):
    plt.scatter(-node['X']+3.05, -node['Y']+1.53, color='red')  # Use red color for nodes

def main():
    plt.figure(figsize=(15, 10))

    for _, arc in arcs.iterrows():
        plot_arc(arc)
    
    for _, node in nodes.iterrows():
        plot_node(node)
        
    for _, segment in segments.iterrows():
        plot_segment(segment)
        

    plt.axis('equal')
    plt.title("Road System Visualization")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.show()

main()
```
